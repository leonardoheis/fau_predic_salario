```python
# Separa cada dataframe en su propio bloque
df_people      = pd.read_csv("../data/people.csv")
df_descriptions = pd.read_csv("../data/descriptions.csv")
df_salary      = pd.read_csv("../data/salary.csv")

# Cargar los datos
file_paths = ['../data/descriptions.csv', '../data/people.csv', '../data/salary.csv']
df_merged = merge_multiple_dataframes(file_paths, 'id')
```

```python
# Unir los tres dataframes por 'id'
#df_merged = df_people.merge(df_salary, on="id").merge(df_descriptions, on="id")

# Eliminar filas con valores nulos
df_merged_clean = df_merged.dropna().reset_index(drop=True)

# Crear IDs categóricos para job title y education level
df_merged_clean['job_title_id'] = df_merged_clean['Job Title'].astype('category').cat.codes
df_merged_clean['education_id'] = df_merged_clean['Education Level'].astype('category').cat.codes

# Crear tablas auxiliares
df_job_titles = df_merged_clean[['job_title_id', 'Job Title']].drop_duplicates().reset_index(drop=True)
df_education = df_merged_clean[['education_id', 'Education Level']].drop_duplicates().reset_index(drop=True)

# Crear EntitySet
es = ft.EntitySet(id="employee_data")

# Agregar dataframes
es = es.add_dataframe(dataframe_name="people", dataframe=df_merged_clean, index="id")
es = es.add_dataframe(dataframe_name="job_titles", dataframe=df_job_titles, index="job_title_id")
es = es.add_dataframe(dataframe_name="education_levels", dataframe=df_education, index="education_id")

# Definir relaciones
rel_job = ft.Relationship(es, "job_titles", "job_title_id", "people", "job_title_id")
rel_edu = ft.Relationship(es, "education_levels", "education_id", "people", "education_id")
es = es.add_relationships([rel_job, rel_edu])

# Aplicar Deep Feature Synthesis
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="people",
    agg_primitives=["mean"],  # Solo promedio
    trans_primitives=[],      # Ninguna transformación adicional
    verbose=True
)

# Guardar o visualizar resultados
#feature_matrix.to_csv("../data/feature_matrix.csv", index=False)
print(feature_matrix.head())
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing   import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df_merged.isnull().sum()
```

```python
df_merged_numerics = df_merged.copy()

numeric_vars = df_merged_numerics.select_dtypes(
    include="number").columns.to_list()
numeric_vars

imputer = SimpleImputer(strategy="median")

# Indicate which variables to impute:
ct = ColumnTransformer(
    [("imputer", imputer, numeric_vars)],
    remainder="passthrough",
    force_int_remainder_cols=False,
).set_output(transform="pandas")

ct.fit(df_merged_numerics)

ct.named_transformers_.imputer.statistics_

df_merged_clean_n = ct.transform(df_merged_numerics)

df_merged_clean_n.isnull().sum()
```

```python
df_merged_categoric = df_merged.copy()

categorical_vars = df_merged_categoric.select_dtypes(
    include="O").columns.to_list()

imputation_dict = {var:
     "no_data" for var in categorical_vars}

imputer = SimpleImputer(strategy='most_frequent')

ct = ColumnTransformer(
    [("imputer",imputer, categorical_vars)],
    remainder="passthrough"
    ).set_output(transform="pandas")

ct.fit(df_merged_categoric)

ct.named_transformers_.imputer.statistics_

df_merged_clean_c = ct.transform(df_merged_categoric)

df_merged_clean_c.isnull().sum()
```

```python
df_merged_cn = df_merged.copy()

preproc = ColumnTransformer(
    [
      ("num", SimpleImputer(strategy="median"),        numeric_vars),
      ("cat", SimpleImputer(strategy="most_frequent"), categorical_vars),
    ],
    remainder="drop",
    verbose_feature_names_out=False   # preserva nombres originales
).set_output(transform="pandas")

# preproc.fit(df_merged_categoric)

# df_final = preproc.transform(df_merged_cn)

df_final = preproc.fit_transform(df_merged_cn)

df_final.isnull().sum()
```

```python
desc = df_final['Description'].copy()
job  = df_final['Job Title'].copy()

df_final_OneHotEncoder = df_final.drop(columns=['Description', 'Job Title'])

categorical_vars = df_final_OneHotEncoder.select_dtypes(include="object").columns.tolist()

ct_ohe = ColumnTransformer(
    transformers=[
        (
            "ohe",
            OneHotEncoder(
                drop=None,
                sparse_output=False,
                handle_unknown='ignore'
            ),
            categorical_vars
        )
    ],
    remainder="passthrough",
    verbose_feature_names_out=False
).set_output(transform="pandas")

df_ohe = ct_ohe.fit_transform(df_final_OneHotEncoder)

df_ohe['Description'] = desc.values
df_ohe['Job Title']   = job.values

#df_final_encoded = df_ohe[original_cols]

print(df_ohe.shape)
print(df_ohe.columns.tolist())
df_ohe
```
